### Largest image size
Images have different sizes.
I cannot load the whole dataset or even 60% of it due to memory shortage, so I have to use a loop.

```python
ds = tfds.load('malaria', 
               split='train', 
               shuffle_files=False,
              data_dir='./data',
               batch_size=1,
              download=False,
              as_supervised=True,
              with_info=False)
print(ds)
```

```python
max1 = 0
max2 = 0
for image, label in ds:
    if image.shape[1] > max1:
        max1 = image.shape[1]
    if image.shape[2] > max2:
        max2 = image.shape[2]
print(max1, max2)
```
